```python
fname = "link_graph_series.txt"
```

```python
with open(fname) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.split() for x in content] 
```

```python
links = []
pages = set ()
for i in range (len (content)):
    links.append ((content[i][0], content[i ][1]))
    pages.add (content[i][0])
    pages.add (content[i][1])

```

```python
page_links = []
```

```python
from collections import defaultdict
```

```python
page_links_to = defaultdict(list)
page_links_from = defaultdict(list)
for l in links:
        page_links_to[l[0]].append (l[1])
        page_links_from[l[1]].append (l[0])
```

```python
pages_h = defaultdict (None)
pages_a = defaultdict (None)

for p in pages:
    pages_h[p] = 1
    pages_a[p] = 1

```

```python
for epoch in range (10):
    temp_pages_h = {}
    temp_pages_a = {}
    for p in pages:
        summa = 0
        for y in page_links_to[p]:
            summa += pages_a[y]
        temp_pages_h[p] = summa
        summa = 0
        for x in page_links_from[p]:
            summa += pages_h[x]
        temp_pages_a[p] = summa
    pages_h = temp_pages_h.copy ()
    pages_a = temp_pages_a.copy ()
```

```python
max_h = 0
max_a = 0
for p in pages:
    if pages_h[p] > max_h:
        max_h = pages_h[p]
        max_p_h = p
    if pages_a[p] > max_a:
        max_a = pages_a[p]
        max_p_a = p
```
